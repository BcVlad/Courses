```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import math
import pylab
import random
```

```python
# функция теоретической плотности распределения
def eps_pdf(l, x):
    return (l * math.exp(-l * x))
```

```python
# функция формирования выборки экспоненциального распределения 
def expon(value,l, n):
    mu = 1/l
    list_x = []
    for i in range(value):
        summa = 0
        for j in range(n):
            summa += mu**2*random.expovariate(mu)
        list_x.append(summa/n)
    return list_x
exponent = expon(1000,0.0125,1)
```

```python
# формируем теоретическую плотность распределения
X = np.arange(0,450,0.45)
Y = [eps_pdf(0.0125, x = i) for i in X]
```

# ФОРМИРУЕМ ГРАФИКИ ТЕОРЕТИЧЕСКОЙ ПЛОТНОСТИ РАСПРЕДЕЛЕНИЯ И НАШЕЙ ЭКСПЕРИМЕНТАЛЬНОЙ ВЫБОРКИ

```python
fig, axes = plt.subplots(figsize=(8,4), dpi=80)
plt.hist(exponent, bins = 'auto', color='purple', normed = True, label='Выборка')



plt.plot(X, Y, label='Tеоретическая плотность', color='g', linestyle='-', lw=4)
plt.xlabel('ЗНАЧЕНИЕ')
plt.ylabel('ВЕРОЯТНОСТЬ')
plt.legend()
plt.show()
```

# ПРОВЕРКА ЦПТ

```python
# Теоретическая плотность нормального распределения
def normal_pdf(x, mu, sigma):
    sqrt_two_pi = math.sqrt(2 * math.pi)
    return (math.exp(-(x-mu) ** 2 / 2 / sigma ** 2)/(sqrt_two_pi * sigma)) 
```

```python
# функция отрисовки для этого пунка работы

def graph_sample_theory(list_x):
#   гистограмма выборки
    fig, axes = plt.subplots(figsize=(8,4), dpi=80)
    plt.hist(list_x,bins = 'auto', color='purple',density = True)


#   график плотности
    plt.plot(sorted(list_x),[normal_pdf(x, np.mean(list_x), np.std(list_x)) for x in sorted(list_x)], color='green', lw=4)
    plt.xlabel('Значение')
    plt.ylabel('Вероятность')
    plt.show()
```

```python
choice = [5,10,50]
for n in choice:
    list_x = expon(1000, 0.0125, n)
    graph_sample_theory(list_x)
```
